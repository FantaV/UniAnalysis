```python
# Import the library to fetch the data from Yahoo!Finance
import yfinance as yf

# Import matplotlib and set the style for plotting
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('classic')
```

```python
# Fetch the futures data
futures_data = yf.download("HE=F", start="2017-01-01", end="2021-05-01")

# Plot the adjusted close price
(futures_data['Adj Close']).plot(figsize=(15, 7))

# Set title and labels for the plot
plt.title('Lean Hog Continuous Futures Adjusted Close Price', fontsize=14)
plt.xlabel('Year-Month', fontsize=12)
plt.ylabel('Price', fontsize=12)
plt.show()  
```
