```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('classic')
import yfinance as yf
import pandas as pd

# Ignore warnings for future deprecation
import warnings
warnings.filterwarnings("ignore")
```

```python
# path to read the data files

# first contract
sp_dec = yf.download('ESZ24.CME', start='2024-01-01')

# second contract
sp_mar =yf.download('ESH25.CME', start='2024-01-01')

# Plot both contracts' Adjusted Close prices
sp_dec['Adj Close'].plot(label='ESZ24 (Dec)', color='orange')
sp_mar['Adj Close'].plot(label='ESH25 (Mar)', color='green')

plt.title('S&P500 December Futures Adjusted Close Price', fontsize=14)
plt.xlabel('Year-Month', fontsize=12)
plt.ylabel('Price', fontsize=12)

# adding the legend
plt.legend()

# show the plot
plt.show()

```

```python
# Download data
sp_dec = yf.download('ESZ24.CME', start='2024-01-01')
sp_mar = yf.download('ESH25.CME', start='2024-01-01')

# Determine the expiration date of the December contract
# S&P 500 futures typically expire on the third Friday of the contract month
dec_expiry = pd.Timestamp('2024-12-20')  # Third Friday of December 2024

plt.figure(figsize=(15, 7))

# Plot December contract up to expiration
plt.plot(sp_dec.loc[:dec_expiry, 'Adj Close'], label='ESZ24 (Dec)', color='orange')

# Plot March contract after December contract expires
plt.plot(sp_mar.loc[dec_expiry:, 'Adj Close'], label='ESH25 (Mar)', color='green')

plt.title('S&P 500 Futures Adjusted Close Price with Contract Rollover', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Price', fontsize=12)

# Adding the legend on the left side
plt.legend(loc='upper left')

# Show the plot
plt.show()
```
