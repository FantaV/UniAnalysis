```python
import pandas as pd
import matplotlib.pyplot as plt 
import yfinance as yf

%matplotlib inline
plt.style.use('_mpl-gallery')
```

```python
# define tickers list
tickers_list = ['AAPL','AMZN','MSFT']

# download data and extract Adj Close colums
price_data = yf.download(tickers_list, start='2024-01-01')['Adj Close']

# set the index to a data time object
price_data.index = pd.to_datetime(price_data.index)

# display
price_data.head()
```

```python
# plot the absolute price series 
# in this way we're able to visualize the datasets compressed on the graphic

plt.figure(figsize=(10,7))

(price_data['AAPL']/price_data['AAPL'].iloc[0]).plot(color='orange')
(price_data['AMZN']/price_data['AMZN'].iloc[0]).plot(color='blue')
(price_data['MSFT']/price_data['MSFT'].iloc[0]).plot(color='green')

# set title and axis label
plt.title('Price in Change',fontsize=14)
plt.xlabel('Year-Month',fontsize=12)
plt.ylabel('Price Change',fontsize=12)
plt.legend()

plt.show()
```

```python
# data of assets costituing the S&P 500
# we can read this with the method 'read_html(url)' of pandas 

# url of the source
url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'

# read and print the stock tickers that make up S&P500
tickers_500 = pd.read_html(url)[0]

tickers_500.head()
```

```python
# convert colums of a dataframe into a list using DataFrame[colums].tolist()

# converting the ticker 'MMM'column to list
ticker_symbol = tickers_500['Symbol'].tolist() 

# clean the symbols
ticker_symbol = [ticker.replace('.','-')for ticker in ticker_symbol]

# get data for this tickers from yfinance
data = yf.download(ticker_symbol,'2021-01-01', auto_adjust=True)['Close']

# show 
data.head()
```

```python
plt.figure(figsize=(10,7))

(data[:]/price_data.iloc[0]).plot()
```
