```python
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline 
import yfinance as yf
```

```python
price_data_cdprojekt = yf.download('CDR.WA', start='2022-01-01', auto_adjust=True)

# time set
price_data_cdprojekt.index = pd.to_datetime(price_data_cdprojekt.index)

# display firts row
price_data_cdprojekt.tail()

(price_data_cdprojekt['Close']).plot(figsize=(10,7))

plt.title('CDPROJEKT Adjusted Close Price', fontsize=14)
plt.xlabel('Year-Month', fontsize=12)
plt.ylabel('Price', fontsize=12)
plt.show()
```
