```python
#  We can download data even from WORLD BANK'S DATABASEwith "wbgapi"

import wbgapi as wb
import numpy as np 
import matplotlib.pyplot as plt

ny_gdp = wb.data.DataFrame(['ny.gdp.mktp.cd'], 
                  mrv=1,                    
                  labels=True).dropna().sort_values(by='NY.GDP.MKTP.CD',
                                                    ascending=False)[:10]
ny_gdp

# CODE: Series code, NY.GDP.MKTP.CD for GDP (current USD)
# MVP = most recent value
# LABELS: Adds the country name to the dataframe when True
```

```python
# GDP OF 6 COUNTRIES IN TRILLION DOLLARD USD 

GDP =  wb.data.DataFrame(['ny.gdp.mktp.cd'],
                         ['USA','CHN','JPN','DEU','IND','GBR','ITA']).T.dropna()/(10**12)  
# T = ".trasponse()" Trasposizione del dataframe e serve per scambiare le righe con le colonne 
# dropna = elimina le righe o colonne con valori mancanti. utile per evitare che il grafico non venga tracciato correttamente mantenendo i dati dati "puliti" per evitare errori di calcolo successivi

GDP = GDP[GDP.index >= '2000']

# convert GDP data into numpy arrays
GDP_values= GDP.values
xtick = np.array([int(x[-4:])for x in GDP.index.tolist()])

# plot the GDP
fig, ax = plt.subplots(figsize=(15,7))

# Plot the series 
for country_index in range(GDP_values.shape[1]):
    ax.plot(xtick, GDP_values[:, country_index], label=GDP.columns[country_index])

# Set the title and axis label
plt.title('GDP', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Trillions of Dollars', fontsize=12)
plt.legend()

# Show the plot
plt.show()
```
